## Ag3

```python
ag3 = malariagen_data.Ag3(
    "simplecache::gs://vo_agam_release_master_us_central1",
    simplecache=dict(cache_storage="../gcs_cache"),
    debug=False,
)
ag3
```

```python
ag3.plot_samples_bar(x="country", sample_sets="3.0")
```

```python
ag3.plot_samples_bar(x="country", color="taxon", sample_sets="3.0")
```

```python
ag3.plot_samples_bar(x="year", color="country", sample_sets="3.0")
```

```python
ag3.plot_samples_interactive_map(
    sample_sets=["3.0"],
    sample_query="year > 2009",
)
```

```python
# Demo using a different basemap provider
# - The map background will appear grey if provision of tiles fails
tile_URL_pattern = "https://tile.openstreetmap.org/{z}/{x}/{y}.png"
ag3.plot_samples_interactive_map(
    sample_sets=["3.0"],
    basemap=ipyleaflet.TileLayer(url=tile_URL_pattern),
)
```

```python
# See the available basemap abbreviations
malariagen_data.anoph.map_params.basemap_abbrevs.keys()
```

```python
# Demo using a basemap abbreviation, case-insensitive
ag3.plot_samples_interactive_map(
    sample_sets=["3.0"],
    basemap="mapNik",
)
```

```python
ag3.plot_sample_location_mapbox(
  color='country',
  sample_sets=["3.0"],
  sample_query="year > 2012",
  zoom=2
)
```

```python
ag3.plot_sample_location_geo(
  color='country',
  sample_sets=["3.0"],
  sample_query="year > 2012",
)
```

## Af1

```python
af1 = malariagen_data.Af1(
    "simplecache::gs://vo_afun_release_master_us_central1",
    simplecache=dict(cache_storage="../gcs_cache"),
    debug=False,
    pre=True,
)
af1
```

```python
af1.plot_samples_bar(x="country", sample_sets="1.0")
```
